# Importing Libraries


```python
import cv2
import matplotlib.pyplot as plt
```

# LOAD THE DATA




```python
#cv2. imread() method loads an image from the specified & the required file.
#The imshow() function in pyplot module of matplotlib library is used to display data as an image
image=cv2.imread("iron man")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# CONVERSION OF COLOUR IMAGE TO GRAYSCALE IMAGE

```python
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(grey_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
#Inversion of the image
invert=cv2.bitwise_not(grey_image)
plt.imshow(cv2.cvtColor(invert, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
#Blur the image 
blur=cv2.GaussianBlur(invert,(21,21),0)
```

```python
#Invert the blurred image
invertedblur=cv2.bitwise_not(blur)
```

```python
sketch=cv2.divide(grey_image,invertedblur,scale=256.0)
```

```python
cv2.imwrite("sketch.png",sketch)
```

```python
plt.imshow(cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('the image is converted into sketch')
plt.show()
```
